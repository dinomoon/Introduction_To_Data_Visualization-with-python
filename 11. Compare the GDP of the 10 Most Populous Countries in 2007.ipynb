```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('gapminder.tsv', sep='\t')
data.head()
```

```python
data_2007 = data[data.year == 2007]
data_2007.head()
```

```python
top10_2007 = data_2007.sort_values('pop', ascending=False).head(10)
```

```python
top10_2007.head()
```

```python
top10_2007["GDP"] = top10_2007['pop'] * top10_2007['gdpPercap']
```

```python
top10_2007.head()
```

```python
x = range(10)
plt.subplot(211)
plt.bar(x, top10_2007['pop'] / 10**6)
plt.xticks([],[])
plt.legend(["Population in Millions"])
plt.title("10 Most Population Countries")

plt.subplot(212)
plt.bar(x, top10_2007['GDP'] / 10**9)
plt.xticks(x, top10_2007.country, rotation = '40')
plt.legend(["GDP in Billions"])
plt.show()
```
